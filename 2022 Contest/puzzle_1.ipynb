```python
# get input data

with open ('input.txt') as f:
    data = np.array([line.strip() for line in f])
            
data
```

```python
idx = np.where(data == '')[0]  # find the blanks
cals = [list(filter(None,array)) for array in np.array_split(data, idx)] #create a list of lists using the indexes of the blanks and remove the blanks
cals
```

```python
elf = []
for array in cals:  #iterate over the list of arrays
    elf.append(sum([int(i) for i in array]))  # sum each array of numbers
most_cals = np.max(elf) # max value in the list - reminder np.argmax(elf)+1 will return the elf number
most_cals
```

```python
# find top 3 elves and their combined calorie count
elf.sort(reverse = True)
elf[0:5]
```
