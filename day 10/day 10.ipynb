```python
# part 2 - complete the missing sequence for the incomplete lines

def get_data():
    os.getcwd()
    os.chdir(r'C:\Users\justi\Github\advent_of_code\day 10')
    
    with open('input.txt') as f:
        lines = f.readlines()

    lines = [x[:-1] for x in lines]  # remove the '\n'
    return lines

def open_backets(string='', row=0):
    brackets = {'(': ')', '[': ']', '<': '>', '{': '}'}
    bracket_score = {')':3, ']': 57, '}': 1197, '>': 25137}
    symbol_match = []
    
    drop = False
    for symbol in string:
        if symbol in brackets.keys():
            symbol_match.append(symbol)
        elif symbol in brackets.values():
            open_bracket = [key for key,value  in brackets.items() if value == symbol][0]
            if  open_bracket == symbol_match[-1]:
                symbol_match.pop()
            else:
                drop = True
                break
            
    return drop

# remove rows from part 1
def remove_rows(data, to_drop):
    data = [row for idx, row in enumerate(data) if idx not in to_drop]
    return data

def complete_squence(string):
    score = 0
    # go line and line and figure out what is missing
    brackets = {'(': ')', '[': ']', '<': '>', '{': '}'}
    bracket_score = {')':1, ']':2, '}':3, '>':4}
    symbol_match = []
    comp_seq = []
    for symbol in string:
        if symbol in brackets.keys():
            symbol_match.append(symbol)
        elif symbol in brackets.values():
            open_bracket = [key for key,value  in brackets.items() if value == symbol][0]
            if  open_bracket == symbol_match[-1]:
                symbol_match.pop()
    #print(symbol_match)
    for symbol in symbol_match[::-1]:
        comp_seq.append(brackets[symbol])
        score *= 5
        score += bracket_score[brackets[symbol]]
    comp_seq
    #print(comp_seq)
    return comp_seq, score
        
        
    
    return

def main():
    data = get_data()
    to_drop = []
    score = []
    for row, string in enumerate(data):
        drop = open_backets(string, row)
        if drop:
            to_drop.append(row)
    data = remove_rows(data,to_drop)
    sequences = []
    for string in data:
        new_seq, line_score = complete_squence(string)
        sequences.append(new_seq)
        score.append(line_score)
    output = dict(zip(score, sequences))
    for k,v in output.items():
        print(f'{v} - {k} total points')
    score.sort()
    print(np.median(score))
    return 



if __name__ == '__main__':
    main()
```
