```python
input = pd.read_csv('input_day_2.csv')
input.info()
```

```python
input = input.iloc[:,0].str.split()
```

```python

moves = [i[0] for i in input]
position = [i[1] for i in input]

```

```python
h = 0
d = 0
aim = 0

for x,i in enumerate(moves):
    if i == 'forward':
        h = h + position[x]
        if aim > 0 :
            d = d + (aim * position[x])
    if i == 'down':
        #d = d + position[x]
        aim = aim + position[x]
    elif i == 'up':
        #d = d - position[x]
        aim = aim - position[x]
    print(i, position[x], 'hoirzontal: %s' % h, 'depth: %s' % d, 'aim: %s' % aim)
```
